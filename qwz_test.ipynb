```python
from functions import bhz_ham, chern_marker
from qwz import qwz_hamiltonian

from koala import example_graphs as eg
from koala.voronization import generate_lattice
from koala import plotting as pl
from koala.lattice import Lattice

import numpy as np
from scipy import linalg as la
from matplotlib import pyplot as plt
```

```python
size = 20
square_lat = eg.square_lattice(size, size)

h = qwz_hamiltonian(square_lat, u_vals=1, h_field=0)
e, v = la.eigh(h)
p = v @ np.diag(e < 0) @ v.conj().T
```

```python
c = chern_marker(square_lat, p, True)
plt.plot(c)
```
